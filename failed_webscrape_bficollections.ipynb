```python
import requests
from bs4 import BeautifulSoup
url = "https://collections-search.bfi.org.uk/web/search/simple"

# To scrape the database, we need to put some information into the search and generate results
# I want to use the expert search eventually, but it was easier to start with the simple search
# As a test, I will try the first search with a query that returns less results:
    # 'search in Film and Television Works'
    # keywords: 'lotte reiniger'

data = {
    'Fields[0].Value': 'lotte reiniger',
    'DatabaseChoices[0].Selected': 'true'
    }
response = requests.post(url, data=data)

soup = BeautifulSoup(response.text, 'html.parser')

print(soup)

# When I was unable to generate any data, I printed the variable 'soup' to see the webpage that was being generated by my code
# The resulting output is just the BFI's error page, which reads "Sorry, there's a problem. The site is still in development, and changes regularly, so please try again later if this error persists."

# I used the Developer's window to read through the code at the url, and experimented with refining the input data for BeautifulSoup
# The output was always the same error page

# This error page is identical to the one you encounter in a web browser on the site when the database freezes or goes down
    # the BFI database frequently crashes and generates this error page in its browser format
# I couldn't discern whether the error message resulted from issues with my code, or was internal to the database

# Whie this code is short, it was the result of hours of testing, reading and revision. So I resolved to manually export data in order to be able to move forward with this project and try to succeed elsewhere
    # I had hoped to return to this code at a later date, but ran out of time. I considered using Selenium in the future rather than requests
```
